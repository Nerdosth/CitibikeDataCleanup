```python
df_reduced = df[['start station name','start station latitude','start station longitude','starttime','end station name','end station latitude', 'end station longitude','stoptime','usertype']]
```

```python
df1_reduced = df1[['start_station_name','start_lat','start_lng','started_at','end_station_name','end_lat','end_lng','ended_at','member_casual']]
```

```python
#changing the names as a personal preference. 

df_rename = df_reduced.rename(columns={'start station name':'start_station_name','start station latitude':'start_lat','start station longitude':'start_lng','starttime':'ride_time_start','end station name':'end_station_name', 'end station latitude':'end_lat','end station longitude':'end_lng','stoptime':'ride_time_end', 'usertype':'usertype'})
```

```python
#changing the names as a personal preference. 
df1_rename = df1_reduced.rename(columns={'start_station_name':'start_station_name','start_lat':'start_lat','start_lng':'start_lng','started_at':'ride_time_start','end_station_name':'end_station_name', 'end_lat':'end_lat','end_lng':'end_lng','ended_at':'ride_time_end', 'member_casual':'usertype'})
```

```python
#Using concat to combine the two dataframes together instead of trying to merge. 
df2 = pd.concat([df_rename, df1_rename], ignore_index=True)
```

```python
#the ride_time_start and ride_time_end were objects and I'm using pd.to_datetime to convert them to a datetime. 
df2['ride_time_start'] = pd.to_datetime(df2['ride_time_start'])
df2['ride_time_end'] = pd.to_datetime(df2['ride_time_end'])
```

```python
#CitiBike changed the name of the users from member and I'm converting all the names to my preferred title of Subscriber and Cusrtomer 
df2['usertype'] = df2['usertype'].replace(['member', 'casual'], ['Subscriber', 'Customer'])
```

```python
#The data is missing some coordinates and station locations and I'm removing those enteries before exporting dataframe to a csv file. 
df3 = df2.dropna()
```

```python
#calculating the ride time that will be used to determine how much time a a subscribing user rides vs a one-time customer. 
df3['ride_time_min'] =  df3['ride_time_end'] - df3['ride_time_start']
df3['ride_time_min'] = df3['ride_time_min'].dt.total_seconds().div(60).astype(int)
df3
```

```python
#removing outliers that would skew the ride time data. 
index_ride_min = df3[ (df3['ride_time_min'] >= 180) | (df3['ride_time_min'] <= 2) ].index
df3.drop(index_ride_min , inplace=True)
df3
```

```python
#creating a helper column that will be used to calculate the total trips by particular paremeters in Tableau. 
df3 = df3.assign(trip_count=1)
```

```python
#Creating a sample size and sorting the values by the ride start time that will be used to create the initial visuals in Tableau. Using the full dataset causes performance issues and slows down the development time. 

df_sample = df3.head(50000)
df_sample.to_csv('citibiketripsSample.csv')
```
